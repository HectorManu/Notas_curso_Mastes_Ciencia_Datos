## Tratar con valores faltantes

Creamos un nuevo marco de datos llamado **'crops'** y para crear un valor *NaN* usamos **np.nan** importando **numpy**.

```python
import numpy as np
mydata = {'Crop':['Rice','Wheat','Barley','Maize'],
         'Yield':[1010,1025.2,1404.2,1241.7],
         'cost':[102,np.nan,20,68]}
crops = pd.DataFrame(mydata)
crops
```

## Tratar con duplicados

Creamos un nuevo marco de datos compuesto por artículos y sus respectivos precios.

```python
data = pd.DataFrame({"Items" : ["TV","Washing Machine","Mobile","TV","TV","Washing Machine"], "Price" : [10000,50000,20000,10000,10000,40000]})
data
```

**duplicated()** devuelve un vector lógico que devuelve *True* cuando se encuentra duplicado.

```python
data.loc[data.duplicated(),:]
data.loc[data.duplicated(keep = "first"),:]
```

De forma predeterminada, **kep = 'first'**, es decir, la primera aparición se considera un valor único y sus repeticiones se consideran duplicados.
Si **keep = "last"**, la última aparición se considera un valor único y todas sus repeticiones se consideran duplicadas.

```python
data.loc[data.duplicated(keep = "last"),:] # las últimas entradas no están allí,
# allí los índices han cambiado.
```

Si **keep = "False"**, entonces considera todas las ocurrencias de las observaciones repetidas como duplicadas.

```python
data.loc[data.duplicated(keep = False),:] # se muestran todos los duplicados
# incluidos los únicos.
```

Para descartar los duplicados, **drop_duplicates** se usa con el valor predeterminado **inplace = False**, keep = 'first' o 'last' o 'False' tiene los significados respectivos con oen duplicated()

```python
data.drop_duplicates(keep = "first")
data.drop_duplicates(keep = "last")
data.drop_duplicates(keep = False,inplace = True)  #by default inplace = False
data
```

## if else en Python

Creamos un nuevo marco de datos del nombre de los estudiantes y sus respectivos signos del zodiaco.

```python
students = pd.DataFrame({'Names': ['John','Mary','Henry','Augustus','Kenny'],
                         'Zodiac Signs': ['Aquarius','Libra','Gemini','Pisces','Virgo']})
```

Las funciones en python se definen usando la palabra clave block **def**, seguida del nombre de la función comonombre del bloque. **La función apply()** aplica la función a lo largo de las filas o columnas del marco de datos.

**Note**: Si usamos 'if else' simple, **debemos cuidar la sangría**. Python no incluye llaves para los bucles y si no.

## Producción

```python
def name(row):
    if row["Names"] in ["John","Henry"]:
        return "yes"
    else:
        return "no"
students['flag'] = students.apply(name, axis=1)
students
```

**Alternativamente**, al importar <numpy> podemos usar **np.where**. El primer argumento es la condición a evaluar, el segundo argumento es el valor si la condición es verdadera y el último argumento define el valor si la condición evaluada devuelve *falsa*

```python
#Esta es una manera más corta en vez de utilizar un if o un else como en el caso anterior
import numpy as np
students['flag'] = np.where(students['Names'].isin(['John','Henry']),'yes','no')
students
```

# Multiples condiciones

```python
# Esta es la menera en la cual se haría inicialmente si no sabemos los numerosos métodos que tiene la libreria de numpy

def mname(row):
    if row["Names"] == "John" and row['Zodiac Signs'] == 'Aquarius':
        return 'yellow'
    elif row["Names"] == "Mary" and row['Zodiac Signs'] == 'Libra':
        return 'blue'
    elif row['Zodiac Signs'] == 'Pisces':
        return 'blue'
    else:
        return 'black'

students['color']=students.apply(mname, axis=1)
students
```

Ahora creamos una lista de condiciones y sus respectivos valores si se evaúa como <True> y usamos **np.select** donde el valor predeterminado es el valor si todas las condiciones son <Falsas>.
Con los conceptor vistos anteriormente.

```python
conditions =[
    (students['Names'] == 'John') & (students['Zodiac Signs'] == 'Aquarius'),
    (students['Names'] == 'Mary') & (students['Zodiac Signs'] == 'Libra'),
    (students['Zodiac Signs'] == 'Pisces')
]

choices = ['yellow','blue','purple']
students['color'] = np.select(conditions, choices, default='black')

students
```

# Funsión o unión sobre la base de una variable común.

Tomamos 2 marcos de datos con diferente número de observaciones:

```python
students = pd.DataFrame({'Names': ['John','Mary','Henry','Augustus','Kenny'],
                         'Zodiac Signs': ['Aquarius','Libra','Gemini','Pisces','Virgo']})
students2 = pd.DataFrame({'Names': ['John','Mary','Henry','Augustus','Kenny'],
                          'Marks' : [50,81,98,25,35]})
```

Usando **pd.merge** podmeos unir los dos marcos de datos. **on = 'Names'** indica que la variable común sobre la base de la cual se combinarán los marcos de datos es 'Nombres'

```python
result = pd.merge(students,students2, on='Names') # Solo toma una 
result
```

De forma predeterminada, **how = "inner"**, por lo tanto, solo toma los elementos comunes en ambos marcos de datos. Si desea todos los elementos en ambos marcos de datos, configure **how = "outer"**


```python
result = pd.merge(students, students2, on = 'Names', how = 'outer') # el solo toma el el resultado de las uniones
result
```

Para tomar solo las intersecciones y todos los valores den la izquerda df establezca **how = 'left**

```python
result = pd.merge(students, students2, on='Names',how = "left")
result
```

De manera similar, **how = 'right'** toma solo intersecciones y todos los valores en right df.

```python
result = pd.merge(students, students2, on='Names',how = "right",indicator = True)
result
```

**indicador = True** crea una columna para indicar si los valores están presentes en ambos marcos de datos o en el marco de datos izquierdo o derecho.
